```python
data.drop_duplicates(subset=['track_id'], keep='first', inplace=True)
```

Note: In this dataset, some of the monthly playlists contained overlapping song. That is, a song would appear in more than one playlist. Since we are dropping duplicates based on the track ID, songs that appear in more than one playlist will get dropped from playlists after the first time it appears.

### Features plotted over time of day

```python
for col in data:
    try:
        data.plot.scatter('month', col, xticks=data['month'])
        data.boxplot(col, 'month')
    except:
        print(col)
```

```python
data.plot('date_added', 'valence')
```

```python
data['time'] = pd.to_datetime(data['date_added'])
```

```python
data.plot(x=data.time.dt.time, y='valence', marker='o', alpha=0.3)
```

```python
data.plot(data.time.dt.hour, y='valence', style='o', alpha=0.3)
```

```python
valence_hour = data.boxplot('valence', data.time.dt.hour)
valence_hour.get_figure().suptitle('')
valence_hour.get_figure().gca().set_xlabel('')
```

Notes:
- looking at the scatterplots, there are clusters of across the time axis, indicating that listening to music tends to happen during specific hours of the day

```python
for col in data:
    try:
        # Scatterplot
        data.plot(data.time.dt.time, y=col, style='o', alpha=0.3)
        
        # Boxplot
        feature_hour = data.boxplot(col, data.time.dt.hour)
        feature_hour.get_figure().suptitle('')
        feature_hour.get_figure().gca().set_xlabel('')
    except:
        print(col)
```
